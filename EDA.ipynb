```python
def categorical_summary(df):
    for col in df.select_dtypes(exclude=["number","bool_","object_"]).columns.tolist():
        if len(df[col].value_counts()) < 1000:
            print(df[col].value_counts())
            print('\n')

    for col in df.select_dtypes(include=["object_"]).columns.tolist():
        if len(df[col].value_counts()) < 1000:
            print(df[col].value_counts())
            print('\n')
```

```python
def drop_na_5percent(df):
    print(len(df))
    threshold = len(df) * 0.05
    cols_to_drop_on = df.columns[df.isna().sum() <= threshold]
    df.dropna(subset = cols_to_drop_on, inplace = True)
    print(len(df))
    return df
```
